```python
x = var('x')
```

```python
halfGraph = plot(ln(1+x)-(1/((1/2)+1)), xmin = -2, xmax= 1.5, ymin = -2,ymax = 2, legend_label = 'k=1/2', color= "blue") ;
thirdGraph = plot(ln(1+x)-(1/((1/3)+1)), xmin = -2, xmax= 1.5, ymin = -2,ymax = 2, legend_label = 'k=1/3', color= "red") ;
fiftyGraph = plot(ln(1+x)-(1/((1/50)+1)), xmin = -2, xmax= 1.5, ymin = -2,ymax = 2, legend_label = 'k=1/50', color= "yellow") ;
hundredGraph = plot(ln(1+x)-(1/((1/100)+1)), xmin = -2, xmax= 1.5, ymin = -2,ymax = 2, legend_label = 'k=1/100', color= "green") ;
```

```python
show(halfGraph + thirdGraph + fiftyGraph + hundredGraph)
```

```python
lnGraph = plot(ln(1+x), xmin = -1, xmax = 1.5, ymin= -2,ymax = 1.5, fillalpha=0.3,legend_label="y = ln(1+x)")
```

```python
func1Graph = plot(x/(1+x), xmin = -1, xmax = 1.5, ymin= -2,ymax = 1.5, fillalpha=0.3,legend_label="y = x/(1+x)", color="orange",title='Inequality 1')
```

```python
func2Graph = plot(1/(1+(x/2)), xmin = -1, xmax = 1.5, ymin= -2,ymax = 1.5, fillalpha=0.3,legend_label="y = 1/(1+(x/2))", color="red",title='Inequality 2')
```

```python
oneGraph = plot(1, xmin = -1, xmax = 1.5, ymin= -2,ymax = 1.5, legend_label="y=1", color="green")
```

```python
lnGraph+oneGraph+func1Graph
```

```python
lnGraph+func2Graph
```

```python
x = 1
```

$$\begin{eqnarray}
inequality\ (1)\  yields\ \frac{1}{2} \le ln(2)\le 1\\
inequality\ (2)\  yields\ \frac{2}{3} \le ln(2)
\end{eqnarray}$$
